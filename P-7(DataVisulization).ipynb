### Simple Imputer

* it is a scikit-learn module for filling missing data

* mean, median,most_frequent, constant


```python
import pandas as pd
import numpy as np

a=np.array([[1,2,np.nan,3,4],
          [10,12,14,15,71],
          [67,98,34,35,90],
          [np.nan,10,23,np.nan,34],
          [67,np.nan,31,54,np.nan],
          [90,np.nan,np.nan,45,90]])
df=pd.DataFrame(a,columns=['a',"b",'c','d','e'])
df
```

```python
from sklearn.impute import SimpleImputer
s=SimpleImputer(strategy='mean')
data=s.fit_transform(df)
data
```

```python
from sklearn.impute import SimpleImputer
s=SimpleImputer(strategy='median')
data=s.fit_transform(df)
data
```

```python
df['b'].median()
```

```python
from sklearn.impute import SimpleImputer
s=SimpleImputer(strategy='most_frequent')
data=s.fit_transform(df)
data
```

```python
from sklearn.impute import SimpleImputer
s=SimpleImputer(strategy='constant',fill_value=0)
data=s.fit_transform(df)
data
```

### subplots:

* it is a function for create multiple plots at a time

* plt.subplots(rows,columns,loc)

```python
plt.subplots(2,2)
plt.show()
```

```python
x=np.array([10,20,30,40])
y=np.array([50,60,70,80])

plt.subplot(2,2,1)
plt.plot(x,y,'r*')

plt.subplot(2,2,2)
plt.plot(x,y,'b-')

plt.subplot(2,2,3)
plt.plot(x,y,'y')

plt.subplot(2,2,4)
plt.plot(x,x**2,'m')

plt.tight_layout(2)
```

```python
# bar plot

plt.bar(x,y,color='g')
plt.show()
```
