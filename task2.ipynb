```python
import pandas as pd

# Load Titanic dataset
df = pd.read_csv("Titanic-Dataset.csv")

# 1. How many passengers died?
num_died = df[df['Survived'] == 0].shape[0]

# 2. How many children were on board (Age < 18)?
num_children = df[df['Age'] < 18].shape[0]

# 3. How many unique passenger classes?
unique_pclasses = df['Pclass'].nunique()

# 4. What percentage of passengers survived?
survival_percentage = df['Survived'].mean() * 100

# 5. How many male and female passengers?
gender_counts = df['Sex'].value_counts()

# 6. What was the average age of passengers?
average_age = df['Age'].mean()

# 7. How many passengers were in each class?
pclass_counts = df['Pclass'].value_counts()

# 8. How many passengers embarked from each port?
embarked_counts = df['Embarked'].value_counts()

# 9. What is the survival rate for each class?
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100

# 10. Survival rate for children vs adults
df['Is_Child'] = df['Age'] < 18
survival_child_vs_adult = df.groupby('Is_Child')['Survived'].mean() * 100

# 11. Class with highest survival rate
highest_survival_class = survival_by_class.idxmax()

# 12. Did gender affect survival chances?
survival_by_gender = df.groupby('Sex')['Survived'].mean() * 100

# 13. Number of families (based on SibSp + Parch > 0)
df['Family_Size'] = df['SibSp'] + df['Parch']
families_traveling_together = df[df['Family_Size'] > 0].shape[0]

# 14. Fare distribution among different classes
fare_distribution = df.groupby('Pclass')['Fare'].describe()

# 15. Average age of survivors vs non-survivors
average_age_survivors = df[df['Survived'] == 1]['Age'].mean()
average_age_nonsurvivors = df[df['Survived'] == 0]['Age'].mean()

# Display results
results = {
    "1_Passengers_Died": num_died,
    "2_Children_On_Board": num_children,
    "3_Unique_Pclasses": unique_pclasses,
    "4_Survival_Percentage": survival_percentage,
    "5_Gender_Counts": gender_counts,
    "6_Average_Age": average_age,
    "7_Pclass_Counts": pclass_counts,
    "8_Embarked_Counts": embarked_counts,
    "9_Survival_By_Class (%)": survival_by_class,
    "10_Survival_Children_vs_Adults (%)": survival_child_vs_adult,
    "11_Highest_Survival_Class": highest_survival_class,
    "12_Survival_By_Gender (%)": survival_by_gender,
    "13_Families_Traveling_Together": families_traveling_together,
    "14_Fare_Distribution_By_Class": fare_distribution,
    "15_Average_Age_Survivors": average_age_survivors,
    "15_Average_Age_NonSurvivors": average_age_nonsurvivors
}

for key, value in results.items():
    print(f"\n{key}:\n{value}")
```
