#### Count plot

```python
help(sns.countplot)
```

```python
sns.countplot(x = "sex",data=df)
```

```python
sns.countplot(x = "size",data = df)
```

```python
#scale and context
sns.set_context("paper",font_scale=2)#poster,paper,notebook,talk
plt.figure(figsize=(20,4))
sns.countplot(x="size",data=df)
```

### Strip plot

```python
help(sns.stripplot)
```

```python
plt.figure(figsize = (10,4))
sns.stripplot(x = "total_bill",data=df)
```

```python
sns.set_style("whitegrid")
sns.stripplot(x = "day",y="tip",hue = "day",data=df)
```

### Swarm plot
- swarmplot is similar to stripplot.but locations of points are changed automatically to avoid the overlapping

```python
sns.set_style("whitegrid")
sns.swarmplot(x = "day",y="tip",hue = "day",data=df)
plt.legend(loc=(1,0.5))
```

```python
sns.swarmplot(x = "total_bill",data=df)
```

```python
sns.swarmplot(x ="time",y="tip",data=df)
```

```python
sns.swarmplot(x = "time",y="tip",data=df,order=["Dinner","Lunch"])
```
