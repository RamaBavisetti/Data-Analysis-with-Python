- z score is also called as standarad score
- zscore=(x-mean)/std.deviation

```python
data = [1,2,2,2,2,3,1,1,1,15,2,2,1,2] #1,2,3...15
mean = np.mean(data)
std = np.std(data)
print(mean)
print(std)
```

```python
t = 3
for i in data:
    z = (i-mean)/std
    if z>t:
        print(i)
    
```
