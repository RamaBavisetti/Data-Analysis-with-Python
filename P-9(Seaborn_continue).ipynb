### Regression Plot:

* Regression plots create a regression line between two features
* It is used for find the linear relationship
* For creating regression plot we can be implement with help of lmplot()[linear model plot] function

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',markers=['d','s'])
```

```python
help(sns.lmplot)
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',markers=['d','s'],palette='Set1')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='smoker',palette=dict(Yes='g',No='m'))
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='sex',row='time',hue='smoker')
```

### joint plot:

* It is a combination of two plots
* Default plot is scatter & Histogram
* Used for relation between two variables

```python
df=sns.load_dataset('iris')
df.sample(5)
```

```python
sns.jointplot(x='sepal_length',y='petal_width',data=df)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='reg')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='kde')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
```

```python
help(sns.jointplot)
```

```python
sns.pairplot(df,hue='species')
```

### HeatMap:

* It is used for find the correlation between features

* correlation shows the relation between two features

* it denotes from -1 to 1

* -1 => it reprasents the negative relation

* 0 => no relation

* 1 => it reprasents the possitive relation


```python
df=sns.load_dataset('iris')
df.sample(5)
```

```python
c=df.corr()
c
```

```python
sns.heatmap(c)
```

```python
sns.heatmap(c,annot=True)
```

```python
sns.heatmap(c,annot=True,linecolor='black',linewidths=2)
```

```python
help(sns.heatmap)
```

```python
#cmap="YlGnBu"
sns.heatmap(c,annot=True,linecolor='black',linewidths=2,cmap="YlGnBu")
```

```python
#cmap='magma'
sns.heatmap(c,annot=True,linecolor='black',linewidths=2,cmap="magma")
```
