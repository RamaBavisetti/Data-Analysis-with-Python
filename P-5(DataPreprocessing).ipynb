### Concatination

```python
d1={"city":["anathapur","Ongole","tirupati","vizag"],
    'count':[100,300,500,200],
   'area':[10,30,50,20]}
df1=pd.DataFrame(d1)
df1
```

```python
d2={'city':['Rajahmundry','Guntur','srikakulam','kadapa'],
   'count':[500,600,350,450],
   'area':[50,60,35,45]}
df2=pd.DataFrame(d2)
df2
```

```python
t=pd.concat([df1,df2])
t
```

```python
t=pd.concat([df1,df2],ignore_index=True)
t
```

```python
t=pd.concat([df1,df2],keys=['a','b'])
t
```

```python
t.loc['a']
```

```python
t=pd.concat([df1,df2],axis=1)
t
```

### Merge/join

```python
d1={"sno":[101,102,103,110],'sname':["a","b",'c','h']}
df1=pd.DataFrame(d1)
df1
```

```python
d2={"sno":[101,102,103,104,105],'saddress':["aa","bb",'cc','dd','ee']}
df2=pd.DataFrame(d2)
df2
```

```python
# default inner join
s=pd.merge(df1,df2,on='sno')
s
```

```python
# outer join
s=pd.merge(df1,df2,on='sno',how='outer')
s
```

```python
# left join
s=pd.merge(df1,df2,on='sno',how='left')
s
```

```python
# right join
s=pd.merge(df1,df2,on='sno',how='right')
s
```

### Data Preprocessing:

* Data preprocessing is used for improve the quality of data

#### Problems in data:

* Insufficient data
* Too much data
* Missing data
* Duplicat data
* Outliers

#### Outliers:

* In general term, it is a data point that is further away from the other data points

* For outliers, we can you use some techniques
    * Standardzation(Standard Scaler)
    * Robot scaling
    * Data Range(MinMaxScaler)
    * Normalization

### Data visulization for outliers:
### KDE (Kernal Dencity Estimate)
* it used for multiple samples data into one graph

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.title("befor standardization data visulization")
sns.kdeplot(df['TV'])
sns.kdeplot(df['radio'])
sns.kdeplot(df['newspaper'])
sns.kdeplot(df['sales'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
s_data=s.fit_transform(df)
s_data
```

```python
s_df=pd.DataFrame(s_data,columns=df.columns)
s_df
```

```python
s_df.describe()
```

```python
plt.title("After standard scaling")
sns.kdeplot(s_df['TV'])
sns.kdeplot(s_df['radio'])
sns.kdeplot(s_df['newspaper'])
sns.kdeplot(s_df['sales'])
plt.show()
```

### Robust scaling:

* it scale using median and IQR(Inter quartile range)


```python
from sklearn.preprocessing import RobustScaler
r=RobustScaler()
r_data=r.fit_transform(df)
r_data
```

```python
r_df=pd.DataFrame(r_data,columns=df.columns)
r_df.head()
```

```python
plt.title("After robust scaling")
sns.kdeplot(r_df['TV'])
sns.kdeplot(r_df['radio'])
sns.kdeplot(r_df['newspaper'])
sns.kdeplot(r_df['sales'])
plt.show()
```

### MinMaxScaler

* Minmax scaling is used for custom range

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler(feature_range=(-1,1))
m_data=m.fit_transform(df)
m_data
```

```python
m_df=pd.DataFrame(m_data,columns=df.columns)
m_df.sample(10)
```

```python
plt.title("After scale in min max scaler")
sns.kdeplot(m_df['TV'])
sns.kdeplot(m_df['radio'])
sns.kdeplot(m_df['newspaper'])
sns.kdeplot(m_df['sales'])
plt.show()
```

### Normalization:

* In certain cases we need to scale the individual data observation(rows)

* when clustering data we need to apply normalization

```python
from sklearn.preprocessing import Normalizer
n=Normalizer()
n_data=n.fit_transform(df)
n_data
```

```python
n_df=pd.DataFrame(n_data,columns=df.columns)
n_df.head()
```

```python
plt.title("After scale using normalization")
sns.kdeplot(n_df['TV'])
sns.kdeplot(n_df['radio'])
sns.kdeplot(n_df['newspaper'])
sns.kdeplot(n_df['sales'])
plt.show()
```
