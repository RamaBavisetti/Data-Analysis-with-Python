```python
import pandas as pd
##4. Get basic statistics for numerical columns

# Load the CSV file
df = pd.read_csv("Students Social Media Addiction.csv")

# Display basic statistics for numerical columns
print(df.describe())
```

```python
# Get the gender distribution
gender_distribution = df['Gender'].value_counts()
gender_distribution

```

```python
# Step-by-step answers to questions 7-20

# 7. Distribution of academic levels
academic_level_dist = df['Academic_Level'].value_counts()

# 8. Average age by academic level
avg_age_by_academic_level = df.groupby('Academic_Level')['Age'].mean()

# 9. Distribution of students by country
country_distribution = df['Country'].value_counts()

# 10. Average daily social media usage by gender
avg_usage_by_gender = df.groupby('Gender')['Avg_Daily_Usage_Hours'].mean()

# 11. Most commonly used social media platforms
most_common_platforms = df['Most_Used_Platform'].value_counts()

# 12. Average addiction score per platform
avg_addiction_by_platform = df.groupby('Most_Used_Platform')['Addicted_Score'].mean().sort_values(ascending=False)

# 13. Top 5 students with highest addiction scores
top_5_addicted_students = df[['Student_ID', 'Addicted_Score']].sort_values(by='Addicted_Score', ascending=False).head(5)

# 14. Count of students who say social media affects academic performance
academic_impact_count = df['Affects_Academic_Performance'].value_counts()

# 15. Average addiction score by academic impact
avg_addiction_by_academic_impact = df.groupby('Affects_Academic_Performance')['Addicted_Score'].mean()

# 16. Correlation between sleep hours and mental health score
sleep_mental_correlation = df['Sleep_Hours_Per_Night'].corr(df['Mental_Health_Score'])

# 17. Compare sleep hours by academic level
avg_sleep_by_academic_level = df.groupby('Academic_Level')['Sleep_Hours_Per_Night'].mean()

# 18. Mental health score vs. daily usage hours (correlation)
mental_vs_usage_correlation = df['Mental_Health_Score'].corr(df['Avg_Daily_Usage_Hours'])

# 19. Average number of conflicts by relationship status
avg_conflicts_by_relationship = df.groupby('Relationship_Status')['Conflicts_Over_Social_Media'].mean()

# 20. Do addicted users report more conflicts?
# Define addicted users as those with score >= 7
df['Is_Addicted'] = df['Addicted_Score'] >= 7
avg_conflicts_by_addiction = df.groupby('Is_Addicted')['Conflicts_Over_Social_Media'].mean()

# Bundle all results into a dictionary to return
results = {
    "7_Academic_Level_Distribution": academic_level_dist,
    "8_Avg_Age_By_Academic_Level": avg_age_by_academic_level,
    "9_Country_Distribution": country_distribution,
    "10_Avg_Usage_By_Gender": avg_usage_by_gender,
    "11_Most_Common_Platforms": most_common_platforms,
    "12_Avg_Addiction_By_Platform": avg_addiction_by_platform,
    "13_Top_5_Addicted_Students": top_5_addicted_students,
    "14_Academic_Impact_Count": academic_impact_count,
    "15_Avg_Addiction_By_Academic_Impact": avg_addiction_by_academic_impact,
    "16_Sleep_Mental_Correlation": sleep_mental_correlation,
    "17_Avg_Sleep_By_Academic_Level": avg_sleep_by_academic_level,
    "18_Mental_vs_Usage_Correlation": mental_vs_usage_correlation,
    "19_Avg_Conflicts_By_Relationship": avg_conflicts_by_relationship,
    "20_Avg_Conflicts_By_Addiction": avg_conflicts_by_addiction
}

results
```
